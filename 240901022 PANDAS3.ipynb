```python
import pandas as pd
import numpy as np
#creating series from array
arr=np.array(['p','a','n','d','a','s'])
a =pd.Series(arr)
print("Series from array:")
print(a)
```

```python
arr={'x':0.,'y':1.,'z':2.}
b=pd.Series(arr)
print("\n\nSeries from dictonary:\n")
print(b)
```

```python
x=pd.Series(4,index=[0,1,2,3])
print("\nSeries using sclar\n")
print(x)
```

```python
x=pd.Series([1,2,3],index=['a','b','c'])
print("\nSeries through index:")
print(x)
```

```python
a=pd.Series(data=[1,2,3,4])
print("\n a series:\n",a)
print("\nIndex:\n",a.index)
print("\nValues:\n",a.values)
print("\nShape:\n",a.shape)
print("\nDiamension:\n",a.ndim)
print("\nSize:\n",a.size)
```
